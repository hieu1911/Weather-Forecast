Do lượng mưa thất thường nên ta sẽ gán lại hôm có mưa là 1 và hôm không mưa là 0

Độ che phủ mây cũng có sự chênh lệnh khá lớn nên ta sẽ gán lại thông số này về 2 khoảng là [0, 50] và [50, 100] tương ứng 2 nhãn là 0 và 1

```python
for i in range(len(dataset_new)):
  if dataset_new['cloudcover'][i] > 50:
    dataset_new['cloudcover'][i] = 1
  else:
    dataset_new['cloudcover'][i] = 0
```

# Seasonal decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
fig = seasonal_decompose(dataset_new['temp'][:730]).plot()
fig.set_size_inches((20, 10))
```

# Unnoise

```python
from scipy.signal import medfilt
```

```python
unnoise_dataset = dataset_new.copy(deep=True)
unnoise_dataset['tempmax'] = medfilt(dataset_new['tempmax'], 3)
unnoise_dataset['tempmin'] = medfilt(dataset_new['tempmin'], 3)
unnoise_dataset['temp'] = medfilt(dataset_new['temp'], 3)
unnoise_dataset['dew'] = medfilt(dataset_new['dew'], 5)
unnoise_dataset['humidity'] = medfilt(dataset_new['humidity'], 3)
unnoise_dataset['windspeed'] = medfilt(dataset_new['windspeed'], 3)
unnoise_dataset['visibility'] = medfilt(dataset_new['visibility'], 3)
```

```python
plt.figure(figsize=(19, 6))
plt.plot(dataset_new['temp'][:360], c='red', label='real')
plt.plot(unnoise_dataset['temp'][:360], c='blue', label='unnoise')
plt.title('Temp real vs unnoise')
plt.legend()
```

```python
fig = seasonal_decompose(unnoise_dataset['temp'][:730]).plot()
fig.set_size_inches((20, 10))
```
